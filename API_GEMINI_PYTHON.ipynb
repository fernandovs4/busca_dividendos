```python
fim = 95000
ini = 0
resposta = ""
while ini < len(text):
    if fim > len(text):
        fim = len(text)
    response = model.generate_content("vou te passar um texto de uma assembleia de uma empresa. Sou acionista dela. Quero que voce extraia as informações mais importantes e me passe um resumo do que foi discutido: " + text[ini:fim])
    resposta += response.text
    ini = fim
    fim += 95000
print(resposta)
```
